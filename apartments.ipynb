```python
import pandas as pd
#Distance
#onsite-> basInfDistanceCenter
#price
#        "adpTerrace": true, ---> Terasa
#resp1 = requests.get('https://cityexpert.rs/api/PropertyView/47019/r')
#response_data = resp1.json()
#print(response_data["onsite"]["basInfDistanceCenter"])
#print(response_data["price"])
#basInfFurnished
price = []
distance = []
square_footage = []
balcony = []
furniture = []

for i in ids:
    resp1 = requests.get('https://cityexpert.rs/api/PropertyView/'+str(i)+'/r')
    response_data = resp1.json()
    price.append(response_data["price"])
    square_footage.append(response_data["size"])
    if response_data["onsite"]["adpTerrace"] == True:
        balcony.append(1)
    else:
        balcony.append(0)
    if response_data["onsite"]["basInfFurnished"] == 1:
        furniture.append(1)
    else:
        furniture.append(0)    
    distance.append(response_data["onsite"]["basInfDistanceCenter"])
df = pd.DataFrame({"price" : price, "distance" : distance, "square_footage" : square_footage, "balcony": balcony, "furniture": furniture})
df.to_csv("apartments.csv", index=False)
```
